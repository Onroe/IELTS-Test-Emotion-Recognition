```python
import cv2
import numpy as np
import math
import os

count =0
capturevid = cv2.VideoCapture('Kopi,Band,8.mp4')
frameRate = capturevid.get(5) #frame rate

try:
    if not os.path.exists('datavideo'):
        os.makedirs('datavideo')
except OSError:
    print ('Error: Creating directory of data')

#Extract one frame per second
while(capturevid.isOpened()):
    frameId = capturevid.get(1) #current frame number
    ret, frame = capturevid.read()
    if (ret != True):
        break
    if (frameId % math.floor(frameRate) == 0):
        filename ="./videoframes/test/frame%d.jpg" % count;count+=1
        if (count % 25 == 0):
         cv2.imwrite(filename, frame)
        
        
capturevid.release()
cv2.destroyAllWindows()
print ("Done!")
```

```python
#Python OOP. NOT USED IN THE FORMULA
import cv2
import os
 
def extractFrames(pathIn, pathOut):
    os.mkdir(pathOut)
 
    cap = cv2.VideoCapture(pathIn)
    count = 0
 
    while (cap.isOpened()):
 
        # Capture frame-by-frame
        ret, frame = cap.read()
 
        if ret == True:
            print('Read %d frame: ' % count, ret)
            cv2.imwrite(os.path.join(pathOut, "frame{:d}.jpg".format(count)), frame)  # save frame as JPEG file
            count += 1
        else:
            break
 
    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
 
def main():
    extractFrames('bigbuckbunny720p5mb.mp4', 'data')
 
if __name__=="__main__":
    main()
```

# Extract frames and average the features in one picture

```python
import cv2
import numpy as np

#loading the video
vidcap = cv2.VideoCapture('Kopi,Band,8.mp4')
success, img = vidcap.read()

# establish a while loop for reading all the video frames
frames = 0
#EDIT THESE VALUES AND copimg VALUES TO ZOOM TO PICTURE
#avg = np.zeros((235,180), dtype=np.float64)
avg = np.zeros((165,180), dtype=np.float64)
while success:

    success, image = vidcap.read()
    #copimg = img[65:300, 0:650]
    #copimg = img[15:250, 300:480]
    copimg = img[15:180,  300:480]
    blur_img = cv2.GaussianBlur(copimg, (5, 5), 3)
    gray_img = cv2.cvtColor(blur_img, cv2.COLOR_BGR2GRAY)        
    avg = np.add(avg, gray_img)

    frames += 1
    
avg = np.divide(avg, frames)
filename ="./videoframes/test/frame_stephen.jpg"
cv2.imwrite(filename, avg)
    
```
